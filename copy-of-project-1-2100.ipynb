```python
import datetime
from collections import defaultdict
import random

class ClothingItem:
    def __init__(self, name, price, description):
        self.name = name
        self.price = price
        self.description = description

    def __str__(self):
        return f"{self.name} - ${self.price:.2f}"

class Customer:
    def __init__(self, first_name, customer_id):
        self.first_name = first_name
        self.customer_id = customer_id

    def __str__(self):
        return f"Customer Name: {self.first_name}, ID: {self.customer_id}"

class Store:
    def __init__(self, name, store_id):
        self.name = name
        self.store_id = store_id
        self.inventory = []
        self.customers = []

    def add_item(self, item):
        self.inventory.append(item)

    def add_customer(self, customer):
        self.customers.append(customer)

class StoreVisit:
    def __init__(self, store, customer, items_purchased):
        self.store = store
        self.customer = customer
        self.items_purchased = items_purchased
        self.entry_time = datetime.datetime.now()

    def __str__(self):
        items_str = ""
        total_price = 0
        for item, quantity in self.items_purchased:
            item_price = item.price * quantity
            items_str += f"  - {item.name} (Quantity: {quantity}, Price: ${item.price:.2f}, Total: ${item_price:.2f})\n"
            total_price += item_price
        return (f"Date & Time: {self.entry_time}\n"
                f"Store Name: {self.store.name}, Store ID: {self.store.store_id}\n"
                f"Customer: {self.customer}\n"
                f"Items Purchased:\n{items_str}"
                f"Total Price: ${total_price:.2f}")

# Create multiple stores
stores = [
    Store("Fashion Hub", 101),
    Store("Style Corner", 102),
    Store("Trendy Threads", 103),
    Store("Chic Boutique", 104),
    Store("Urban Wear", 105)
]

# Add inventory to each store
items = [
    ClothingItem("T-Shirt", 20.00, "Cotton t-shirt"),
    ClothingItem("Jeans", 40.00, "Denim jeans"),
    ClothingItem("Jacket", 60.00, "Leather jacket"),
    ClothingItem("Sweater", 30.00, "Woolen sweater"),
    ClothingItem("Hat", 15.00, "Stylish hat")
]

for store in stores:
    for item in items:
        store.add_item(item)

# Add customers
customers = [
    Customer("Alice", 1),
    Customer("Bob", 2),
    Customer("Charlie", 3),
    Customer("Diana", 4),
    Customer("Eve", 5),
    Customer("Frank", 6),
    Customer("Grace", 7),
    Customer("Hank", 8),
    Customer("Ivy", 9),
    Customer("Jack", 10)
]

for store in stores:
    for customer in customers:
        store.add_customer(customer)

# Record store visits with at least 1 bulk buyer per store
store_visits = [
    StoreVisit(stores[0], customers[0], [(items[0], 2), (items[1], 2)]),
    StoreVisit(stores[0], customers[1], [(items[2], 1), (items[3], 2)]),
    StoreVisit(stores[1], customers[2], [(items[1], 3), (items[3], 1)]),
    StoreVisit(stores[1], customers[3], [(items[4], 4)]),
    StoreVisit(stores[2], customers[4], [(items[0], 2), (items[2], 2)]),
    StoreVisit(stores[2], customers[5], [(items[1], 3)]),
    StoreVisit(stores[3], customers[6], [(items[3], 4)]),
    StoreVisit(stores[3], customers[7], [(items[0], 2), (items[4], 3)]),
    StoreVisit(stores[4], customers[8], [(items[2], 2), (items[3], 3)]),
    StoreVisit(stores[4], customers[9], [(items[1], 4)])
]

# Print store visits
for visit in store_visits:
    print(visit)

# Analyze Sales Data
sales_data = defaultdict(int)  # Item name to count

for visit in store_visits:
    for item, quantity in visit.items_purchased:
        sales_data[item.name] += quantity

# Assign random sales numbers between 25 and 115
for item in sales_data:
    sales_data[item] = random.randint(25, 115)

sorted_sales = sorted(sales_data.items(), key=lambda x: x[1], reverse=True)

print("\nTop Selling Items:")
for item, quantity in sorted_sales:
    print(f"{item}: {quantity}")

# Identify Bulk Buyers
def find_bulk_buyers(visits, min_items=3):
    """Find customers who purchase more than a specified number of items."""
    bulk_buyers = set()
    for visit in visits:
        total_items = sum(quantity for _, quantity in visit.items_purchased)
        if total_items >= min_items:
            bulk_buyers.add(visit.customer)
    return bulk_buyers

# Example usage for bulk buyers
bulk_buyers = find_bulk_buyers(store_visits, min_items=3)

print("\nCustomers who buy more than three items at a time:")
for customer in bulk_buyers:
    print(customer)
```
